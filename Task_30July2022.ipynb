```python
import pandas as pd
```

```python
df1 = pd.read_csv('AgentPerformance (1).xlsx - Agent Performance Report.csv')
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
df1['Week'] = df1['Date'].dt.week
l = df1['Average Response Time'].str.split(":")
df1['Converted Average Response Time'] = (l.str[0].astype(int) * 3600 + l.str[1].astype(int) * 60 + l.str[2].astype(int))
l = df1['Average Resolution Time'].str.split(":")
df1['Converted  Average Resolution Time'] = (l.str[0].astype(int) * 3600 + l.str[1].astype(int) * 60 + l.str[2].astype(int))
df1.dtypes
```

```python
df1_modified = df1[df1['Total Chats'] > 0]
```

```python
#q1 - Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave
df2 =  df1_modified.groupby(['Week','Agent Name'])['Average Rating'].mean()
df2 = df2.to_frame()
df2
```

```python
#Q2 - Total working days for each agents
df3 =  df1_modified.groupby('Agent Name')['Date'].nunique()
df3 = df3.to_frame()
df3
```

```python
#q3 - Total query that you hvae taken 
df5 = df1.groupby('Agent Name')['Total Chats'].sum()
df5 = df5.to_frame()
df5
```

```python
#q4 - total Feedback that you have received
df6 = df1.groupby('Agent Name')['Total Feedback'].sum()
df6 = df6.to_frame()
df6
```

```python
#q5 - agent name who have average rating between 3.5 to 4 6
df7 = df1_modified.groupby('Agent Name')['Average Rating'].mean()
df7 = df7.to_frame()
df7[(df7['Average Rating'] > 3.5) & (df7['Average Rating'] < 4.6)].index
```

```python
#q6 - Agent name who have rating lesss then 3.5
df7[(df7['Average Rating'] < 3.5)].index
```

```python
#q7 - agent name who have rating more then 4.5
df8 = df7[(df7['Average Rating'] > 4.5)].index
df8
```

```python
#q8 - how many feedaback agents have received more then 4.5 average
df9 = df6[df6['Total Feedback'] > 0]
df9['Rating'] = df7['Average Rating']
df9[df9['Rating'] > 4.5]
```

```python
#q9 - average weekly response time for each agent 
df10 = df1_modified.groupby(['Week','Agent Name'])['Converted Average Response Time'].mean() / 60
df10 = df10.to_frame(name = 'Average Weekly Response time in minutes')
df10
```

```python
#q10 - average weekely resolution time for each agents
df11 = df1_modified.groupby(['Week','Agent Name'])['Converted  Average Resolution Time'].mean()/60
df11 = df11.to_frame(name = 'Average Weekly Resolution time in minutes')
df11
```

```python
#q11- ist of all agents name
df1['Agent Name'].unique()
```

```python
#q12 - percentage of chat on which they have received a feedback 
df12 = df1_modified.groupby('Agent Name')['Total Chats'].sum() 
df12 = df12.to_frame()
df12['Total Feedback'] = df1_modified.groupby('Agent Name')['Total Feedback'].sum() 
df12['Chat-feedback Percentage'] = (df12['Total Feedback'] / df12['Total Chats']) * 100
df12
```

```python
#q13 - Total contributation hour for each and every agents weekly basis
df00 = pd.read_csv('Agent_Login_Report (4).xls - Agent Login Report.csv',skiprows=2)
l = df00['Duration'].str.split(":")
df00['hours'] = (l.str[0].astype(int) * 3600 + l.str[1].astype(int) * 60 + l.str[2].astype(int)) / 3600
df00['Date'] = pd.to_datetime(df00['Date'])
df00['Week'] = df00['Date'].dt.week


df13 = df00.groupby(['Week','Agent'])['hours'].sum()
df13 = df13.to_frame(name='hour contributed per week')
df13
```

```python
#q14 - total percentage of active hour for a month
df14 = ((df00.groupby('Agent')['hours'].sum()) / (30*24)) * 100
df14 = df14.to_frame('Active hours percentage')
df14
```
